## Feature Engineering

```python
ipl['win_ratio'] = ipl['win_by_runs']/ipl['win_by_wickets']
ipl.tail(2)
```

```python
import pandas as pd
train = pd.read_csv('./deliveries.CSV')
train.tail()
```

```python
train.head()
```

```python
train['over_ball_ratio'] = train['over'] / train['ball']
train.sample(5)
```

```python
ipl.sample()
```

```python
ipl['is_tall'] = ipl['win_by_runs'].apply(lambda x: 1 if x > 25  else 0)
ipl.tail()
```

```python
train['first_name'] = train['batsman'].apply(lambda x: x.split()[0] if isinstance(x, str) else x)
train.sample(5)
```

```python
train['ball_category'] = pd.cut(
    train['ball'],
    bins=[1,2,3,4,5],
    labels=['FirstOne','MiddleOne','LastOne','NotLast']
)
train.sample(5)
```

## Correlation Analysis

```python
import numpy as np
coller = ipl.select_dtypes(include=np.number)
coller.head()
```

```python
coller.corr()
```

```python
coller.corr()['season'].sort_values(ascending=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
plt.figure(figsize=(8,6))
sns.heatmap(coller.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
corr_matrix = coller.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.9)]
print('Redundant columns to drop:', to_drop)
```
