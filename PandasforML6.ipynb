```python
coffee = pd.read_csv('./coffee.CSV')
coffee.head()
```

##### Group by functions in pandas

```python
coffee.groupby(['Coffee Type'])['Units Sold'].mean()
```

```python
coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
coffee.groupby(['Coffee Type', 'Day']).agg({'Units Sold': 'sum'})
```

##### Pivot Table

```python
pivot = coffee.pivot(columns='Coffee Type', index='Day')
pivot.sum()
```

```python
pivot.sum(axis=1)
```

##### Using DateTime with Groupby

```python
bois['born_date'] = pd.to_datetime(bois['born_date'])
bois['month_born'] = bois['born_date'].dt.month
bois['year_born'] = bois['born_date'].dt.year
bois.groupby([bois['year_born'],bois['month_born']])['name'].count().reset_index().sort_values('name', ascending=False)
```

#### Advanced Functionality

```python
# shift() rank() cumsum() rolling()
```

```python
coffee['yesterday_revenue'] = coffee['Units Sold'].shift(2)
coffee
```

```python
coffee['pct_change'] = coffee['Units Sold'] / coffee['yesterday_revenue']
coffee
```

```python

latte = coffee[coffee['Coffee Type']=="Latte"].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
