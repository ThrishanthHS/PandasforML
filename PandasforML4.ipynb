```python
import pandas as pd
coffee = pd.read_csv('./coffee.CSV')
coffee
```

```python
import numpy as np
coffee['new_price'] = np.where(coffee['Coffee Type']=='Espresso', 3.99, 5.99) 
coffee
```

```python
coffee[['Day', 'Coffee Type','new_price']] # Access specific columns
```

```python
coffee.drop(columns=['new_price'])
```

```python
coffee['revenue'] = coffee['Units Sold']*coffee['new_price']
coffee # Multiplication of two rows
```

```python
coffee.rename(columns={'new_price':'price'}) # to change the name of the columns
```

```python
bois = pd.read_csv('./bios.csv')
bois.head()
```

```python
bois_new = bois.copy()
```

```python
# creates a new column called first_name in the DataFrame bois_new by extracting the first word from the name column
bois_new['first_name'] = bois_new['name'].str.split(' ').str[0]
bois_new
```

```python
bois_new.query('first_name == "Keith"') # access name start from Keith
```

```python
bois_new.info()
```

```python
# converts the born_date column into a proper datetime format and stores it in a new column born_datetime.
bois_new['born_datetime'] = pd.to_datetime(bois_new['born_date'])
bois_new
```

```python
bois_new.info()
```

```python
bois_new['born_year'] = bois_new['born_datetime'].dt.year
bois_new
```

```python
bois_new[['name','born_year']]
```

```python
bois_new.to_csv('./bois.csv', index=False)
```

```python
bois['height_category'] = bois['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else "Tall"))
bois.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'heightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bois['category'] = bois.apply(categorize_athlete, axis=1)

bois.head()
```
