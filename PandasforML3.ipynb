##### Filtering Data in Pandas

```python
import pandas as pd
bois = pd.read_csv('./bios.CSV')
bois.head() # show df from top
```

```python
bois.tail()  #show df from bottom
```

```python
bois.info() # Shows information of the dataset
```

```python
bois.loc[bois["height_cm"]>215]  # access players who is Height>210 using loc
```

```python
bois[bois["height_cm"]>215]  # Without using loc
```

```python
bois.loc[bois["height_cm"]>215,['name','height_cm']] #accees only name and height >215
```

```python
bois[bois["height_cm"]>215 & (bois['born_country']=="USA")] #bois from USA & and height>215
```

```python
bois[bois['name'].str.contains("Thri")] # Name start with Thri
```

```python
bois[bois['name'].str.contains("keith", case=False)] # If case is used as case sensitive
```

```python
bois[bois['name'].str.contains("keith|Patrik", case=False)] # Name starts with keith or patrik
```

```python
# Athlete born in a year start with year 19
bois[bois['born_date'].str.contains(r'^19', na = False)]
```

```python
# end of the string can contain zero or more non-vowel characters
bois[bois['name'].str.contains(r'[^AEIOUaeiou]*$',na = False)]
# [^AEIOUaeiou]*$ string ends with zero or more characters that are NOT vowels. $=end of the string
```

```python
# the person is born in USA, FRA, or GBR and their name starts with “Keith”.
bois[(bois['born_country'].isin(["USA","FRA","GBR"])) & 
     (bois['name'].str.startswith("Keith"))]
```

```python
bois.query('born_country == "USA"') #Player born in usa in simple way
```

```python
# names that contain the same letter repeated consecutively.
bois[bois['name'].str.contains(r'(.)\1', na=False)]
```

```python
# Name that contain son or sen
bois[bois['name'].str.contains(r'son|sen$', case=False, na=False)]
```

```python
# Born year starts with 19
bois[bois['born_date'].str.contains(r'^19', na=False)]
```

```python
# Born city has exactly 7 characters
bois[bois['born_city'].str.contains(r'^.{7}$', na=False)]
```

```python
# Contain 3 or more vowels
bois[bois['name'].str.contains(r'([AEIOUaeiou].*){3,}', na=False)]
```
