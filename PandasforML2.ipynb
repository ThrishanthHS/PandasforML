##### Accessing Data with Pandas

```python
import pandas as pd
coffee = pd.read_csv('./coffee.CSV')
coffee.head()
```

```python
coffee.tail(2)
```

```python
coffee.sample(5) # Random Access
```

```python
coffee.sample(10, random_state=42)  # Specific value
```

```python
display(coffee) # Display complete  data set
```

##### Accessing specific value

```python
coffee.loc[0] # Specific value
```

```python
coffee.loc[[0,1,5]] # Multiple rows
```

```python
coffee[0:3] # To access part of the rows
```

```python
coffee.loc[5:8, ["Day"]] # Accees only days from 5ht to 8th rows
```

```python
coffee.loc[9:, ["Day","Units Sold"]] # Access all days and units sold of all rows, :->all the rows
```

```python
coffee.iloc[:5,[0,2]] # or access column wise through index
```

```python
coffee.index=coffee["Day"]  # changes the index of the df
coffee.loc['Monday']  # Get all data from monday
```

```python
coffee.iloc[1:4, coffee.columns.get_loc("Units Sold")] = 10
coffee.head()
```

```python
coffee["Day"]
```

```python
coffee.Day
```

```python
coffee.sort_values( "Units Sold") # Increasing order
```

```python
coffee.sort_values("Units Sold").iloc[10] # Specific items
```

```python
coffee.iloc[0:5].sort_values("Units Sold") # Only sort first 5 indexs
```

```python
coffee.sort_values( "Units Sold", ascending=False) # Ascending order
```

```python
coffee.sort_values(["Units Sold","Coffee Type"], ascending=[0,1])  # Ascending order of coffee type
```

```python
# Using for loop
for index, row in coffee.iterrows():
    print(index)
    print(row['Units Sold'])
    print("\n")
```
