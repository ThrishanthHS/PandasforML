### Merging and Concatenating data

```python
import pandas as pd
bois = pd.read_csv('./bios.CSV')
bois.head()
```

```python
nocs = pd.read_csv('./noc_regions.CSV')
nocs.head()
```

```python
# combines (joins) two DataFrames using matching country codes
bois_new = pd.merge( bois, nocs, left_on = 'born_country', right_on = 'NOC', how='left')
bois_new.head()
```

```python
bois_new.rename(columns = {'region':'born_country_full'}, inplace=True)
bois_new.head()
```

```python
# filters and shows rows where NOC_x and born_country_full values are different.
bois_new[bois_new['NOC_x'] != bois_new['born_country_full']]
bois_new.head()
```

```python
# It shows the name, NOC_x, and born_country_full for rows where NOC_x and born_country_full are different
bois_new[bois_new['NOC_x'] != bois_new['born_country_full']][['name','NOC_x','born_country_full']]
bois_new.head()
```

```python
# collect rows whos born in usa and gbr
usa = bois[bois['born_country'] == 'USA'].copy()
gbr = bois[bois['born_country'] == 'GBR'].copy()
```

```python
# create new df by combining above two df's
new_df = pd.concat([usa,gbr])
new_df.head()
```

```python
results = pd.read_csv('./results.CSV')
results.head()
```

```python
# joins the results and bois DataFrames using the athlete_id column.
combined_df = pd.merge(results, bois, on='athlete_id',how='left')
combined_df.head()
```

### Handling Null Values

```python
coffee = pd.read_csv('./coffee.CSV')
coffee.head()
```

```python
coffee.info()
```

```python
import numpy as np
coffee.loc[[2,3], 'Units Sold'] = np.nan
coffee
```

```python
coffee.isna().sum()
```

```python
coffee.fillna(1001)
```

```python
# Calculate the mean value values and place the null values
coffee.fillna(coffee['Units Sold'].mean()) 
```

```python
coffee['Units Sold'].interpolate()
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate()
coffee.head()
```

```python
coffee.dropna()
```

```python
coffee[coffee['Units Sold'].notna()]
```
